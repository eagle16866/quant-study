```python
import fix_yahoo_finance as yf
import numpy as np
stock = yf.download("fb", start="2010-01-01", end="2017-06-30")
```

```python
stock.tail()
```

```python
stock['Log_Ret'] = np.log(stock['Close'] / stock['Close'].shift(1))
stock['Volatility'] = stock['Log_Ret'].rolling(window=252).std() * np.sqrt(252)
```

```python
%matplotlib inline
stock[['Close', 'Volatility']].plot(subplots=True, color='blue',
                                   figsize=(8, 6), grid=True);
# tag: goog_vola
# title: Google closing prices and yearly volatility
```

```python
stock[['Adj Close', 'Volatility']].plot(subplots=True, color='blue',
                                   figsize=(8, 6), grid=True);
```

```python
stock['log_ret']=np.log(stock['Adj Close']/stock['Adj Close'].shift(1))
stock['volatility']=stock['log_ret'].rolling(window=252).std() * np.sqrt(252)
```

```python
stock[['log_ret','volatility']].plot(subplots=True,color='Red',figsize=(12,6))
```
